# Lab-02 - Morphological Image Processing


    Goal :  Identify ‘a’ in this picture

## Methodology
1. Thresholding - Ensure a clean binary input.
2. Opening - Remove small pixel noise.
3. Hit-or-Miss - Isolate pixels that match the 'a' structure.
4. Dilation - Expand the detected "points" back into the original letter shape.
5. Validation - Use Area/Centroid checks to confirm the match.
